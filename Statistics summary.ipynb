# HYPOTHESIS TESTING

1. Establish the null (Ho) and the alternative hypothesis (Ha/ H1)
Null hypothesis is made in such a way that if it is found to be true then

a. no action is needed<br>
b. no decision is required<br>
c. no changes are required<br>
d. maintain status quo<br>

Swiggy claims that max time of deliver is 40 min/ travel to CST
Ho t < = 40
Ha t > 40
error is only on one side which is right side
right tail test

Davendar claims that the min salary is 10 LPA/ purchasing rice

Ho sal >= 10<br>
Ha sal < 10<br>
error is only on one side which is left side<br>
left tail test<br>


Ho voltage = 220<br>
Ha voltage != 220<br>
error is on both sides<br>
two tail test<br>


2. Establish the confidence level and level of significance (LoS)
Level of significance = 100 - confidence level
Be default the CL = 95%

3. Based on type of data and type of situation, choose the appropriate statistical test

4. Based on data, calculate the test statistic and the p-value (done by Python)

5. Compare the p-value (from Step#4) against the LoS (from step#2).
IF THE P-VALUE IS LESS THAN THE LOS THEN WE REJECT HO<br>
IF THE P-VALUE IS MORE THAN THE LOS THEN WE DO NOT REJECT HO

**Type-I** : REJECTING a Ho which is true - rejecting an important phone call<br>
**Type-II** : NOT rejecting a Ho which is false - receiving a spam call

p-value (from step#4) is the actual probability of committing the type-I error<br>
LoS (from step#2) is the MAX allowed probability of committing the type-II error


When the error is on the left side (left tail test), we calculate the 95% upper bound because we wish to know the highest value that the sample mean can achieve. For this we ADD z*sigma/sqrt(n) to the sample mean.

So if the claim value is anything between the sample mean and 95 UB, then the null hypothesis is NOT rejected. But if the claim value is higher than the UB then the Ho is rejected.



When the error is on the right side (right tail test), we calculate the 95% lower bound because we wish to know the lowest value that the sample mean can achieve. For this we SUBTRACT z*sigma/sqrt(n) from the sample mean

So if the claim value is anything between the sample mean and 95 LB, then the null hypothesis is NOT rejected. But if the claim value is lower than the LB then the Ho is rejected.

<h3>Test 1</h3>
1-sample Z test/ 1-sample t test (used ONLY for continuous data)
This test is used to compare the sample mean with the claim value and conclude whether the difference between them is statistically significant or not.

z = (Xbar-mu)/(sd/qrt(n))
Xbar- sample mean
mu - claim value
sd - standard deviation
n - sample size

ssw.ztest(x1,value=7,alternative='smaller')

For left tail we use 'smaller'
For right tail we used 'larger'
For two tail test we use 'two-sided'

Example: I wish to check whether the average thickness of the con-rods manufactured is atleast 5.6mm or not.

<h3>Test 2</h3>
2-sample t test (used ONLY for continuous data)
This test is used to compare the difference of the sample means of two independent samples with a claim value and establish whether the difference is statistically significant or not.

Example: We need to check and establish whether the speed of vehicles in Delhi is atleast 10kmph more than that in Blore or not.

For a left tail test
If the claimed value is less than the upper bound then the Ho is NOT rejected
If the claimed value is more than the upper bound then the Ho is rejected

For a right tail test
If the claimed value is more than the lower bound then the Ho is NOT rejected
If the claimed value is less than the lower bound then the Ho is rejected


<h3>Test 3</h3>
Paired t test (used ONLY for continuous data)
Key word: comparison of BEFORE AND AFTER
Used to compare the average difference between the paired values of two samples with the claim value and establish whether the difference is statistically significant or not.
In 2 sample test we compare the difference of the averages against a claim value
In pared t test we compare the average of the differences against a claim value


<h3>Test 4</h3>
One way ANOVA (ANalysis Of VAriance) Used for the continuous data
Used when there are multiple samples (more than 2). This test checks whether the means are equal for all the pairs of samples or not.
Standard Ho: all the pair of samples have equal means
Standard Ha: atleast one of the pair of samples have unequal means

Factor: the variable which can influence the response (input)
Response : the variable getting influnced (output)

factor is who is the teacher - Vineeta or Sudeep or Ruchika
response : marks that the students get.

Anova would check whether the marks depend on the teacher or not.

Assumptions regarding ANOVA
1. All the values are continuous data
2. Normally distributed
3. Equal variance of all samples

<h3>Test 5</h3>
Two way ANOVA Used for the continuous data
Used when the response is getting influenced by two or more categorial factors.

Interaction between two factors is said to occur if the change in the levels of one factor has a different influence on the levels of the other factor.
eg when the weather changes from dry to rain, it has a different impact on the speeds of car and bike.

<h3>Test 6</h3>
1-proportion test (Used for the discrete data)
Used when the proportion calculated from a sample has to be compared against a claim value and check whether the difference between them is statistically significant or not
Eg. I may claim that 30% of people in Kharghar are grads.

<h3>Test 7</h3>
2-proportion test (Used for the discrete data)
Used when the difference of the two proportions calculated from two samples has to be compared against a claim value and establish whether the difference is statistically significant or not.
Eg. The difference of proportion of grads in Mumbai and Blore is max 0.15 or 15%

<h3>Test 8</h3>
Chi square test for the goodness of fit
Used to compare the counts of categories with the claimed proportions and establish whether the counts follow the pattern or not
Standard Ho: the counts follow the pattern
Standard Ha: the counts do not follow the pattern

<h3>Test 9</h3>
Chi square test for degree of association
Used to compare whether the counts obtained for the different levels of one factor depend on the levels of the second factor or not.
eg. whether the choice of the food depends on the gender or not.
whether the choice of the industry depends on the field of graduation or not.

Standard Ho: there is no association between the two factors
Standard Ha: there is an association between the two factors
