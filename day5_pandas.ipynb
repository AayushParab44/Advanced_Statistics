```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='disease')
df.head()
```

```python
df2=df.melt(id_vars=['Country','Disease'],var_name='Year',value_name='Cases')
df2.head()
```

```python
df3=df2.pivot_table('Cases',['Country','Year'],'Disease')
print(type(df3))
df3.head()
```
