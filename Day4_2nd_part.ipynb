```python
sal=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Salary')
```

```python
dept=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Deptt')
```

```python
sal
```

```python
dept
```

```python
pd.merge(sal,dept)  #different implementation  , merging is based on two columns which are commmon in both the dfs
```

```python
pd.merge(sal,dept,how='outer')  #this gives the union of sal and dept
```

```python
pd.merge(sal,dept,how='left')  #left join
```

```python
pd.merge(sal,dept,how='right')  #left join
```
