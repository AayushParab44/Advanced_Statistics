# Which graph to be used where and when is important

#we can plot a discrete data graph and if this graph is considered over a very long period of time then it becomes a continuous graph

- when n*p and n(1-p) > 10   #For a large sample size discrete data becomes continuous
- (ex. when n>10000)
- p- Probabiity
- n= sample size



-- An example is (salary) and (number of covid cases from 2020 to 2024)

Salary can be measured in paise. If each and every point in the salary is plotted then it becomes a discrete graph. \
But if we consider this for a very long duration it becomes a continuous graph(or is very similiar to a continuous graph).


Similarly covid cases were counted for each and every day, but if we plot it for a very long  \
period of time i.e. from 2020 to 2024 we eventually get a continuous graph.

# CATEGORICAL PLOTS-
<h3>1. BAR PLOT</h3>

- The bars are of equal width.  

- The bars will not touch each other(of 2 different categories)

```python
states=['MH','KA','TN','UP','AP','MP']
counts=[23,65,45,70,35,20]
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.bar(states,counts)  #In bar graph width of each and every bar is the same
```

```python
colors=['Violet', 'Indigo', 'Blue', 'Green', 'Yellow', 'Orange', 'Red']
plt.title('States Graph')
plt.xlabel('States')
plt.ylabel('Number')

# plt.figure(figsize=(5,5))
plt.bar(states,counts,color=colors)
```

```python
colors=['Violet', 'Indigo', 'Blue', 'Green', 'Yellow', 'Orange', 'Red']
colors=['#FFD700', '#001F54', '#DC143C', '#228B22', '#008080', '#2F4F4F']
plt.title('States Graph')
plt.xlabel('States')
plt.ylabel('Number')

# plt.figure(figsize=(5,5))
plt.bar(states,counts,color=colors)
```

```python
sns.barplot(x=states,y=counts,palette=('Blues'))
```
