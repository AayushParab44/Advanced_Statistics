# Important for skewedness
- peak on left --> skewed to right --> positively skewed --> skew>0 --> salary
- peak on right --> skewed to left --> negativel skewed --> skew<0 --> parliament
- peak in the center --> symmetric --> skew=0 --> business operations



- peak on left --> mean > median > mode
- peak on right --> mean < median < mode


- formula for skewness = sum((Xi-Xbar)**3)/(n-1)*sd**3)
  xi- each value
  xbar - mean
  sd -standard deviation
  n=

```python
df=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='faithful')
```

```python
erupt=df['eruptions']
```

```python
sum((erupt-np.mean(erupt))**3/((len(erupt)-1)*(np.std(erupt,ddof=1)**3)))  #left skewdness
```

# Kurtosis
- it signifies the peakedness of the graph.
- it shows the sharpness of the peak.

- formula for excess kurtosis = sum((Xi-Xbar)**4)/(n-1)*sd**4)
- formula for kurtosis = sum((Xi-Xbar)**4)/(n-1)*sd**4) - 3  

- more peaked than normal - kurtosis > 0
- less peaked than normal - kurtosis < 0
  

```python
# median is one moment above the mean
# median is second moment above the mean
# skewness is third moment above the mean
# kurtosis is fourth moment above the mean
```
